```python
import pandas as pd
import datetime
sales_data = pd.read_csv('https://raw.githubusercontent.com/mahanteshimath/LEARN-DATAMODELING/refs/heads/main/src/sales_data.csv')
sales_data.head(10)
```

```python
import datetime
from datetime import datetime, timedelta
sales_data['timestamp'] = pd.to_datetime(sales_data['timestamp'])
region_to_analyze = 'North'
end_date = datetime.now()
start_date = end_date - timedelta(days=6 * 30)  

filtered_data = sales_data[
    (sales_data['region'] == region_to_analyze) &
    (sales_data['timestamp'] >= start_date) &
    (sales_data['timestamp'] <= end_date)
]
filtered_data.head(10)
```

```python
product_sales = (
    filtered_data.groupby('product_id')['sales']
    .sum()
    .reset_index()
    .sort_values(by='sales', ascending=False)
)
top_3_products = product_sales.head(3)
print("Top 3 Best-Selling Products in the Region:", region_to_analyze)
product_sales.head(3)
```
